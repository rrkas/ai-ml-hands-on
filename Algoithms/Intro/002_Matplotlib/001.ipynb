```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(5, 100, 100)
y = np.linspace(10, 1000, 100)

plt.plot(x, y)
plt.show()
```

```python
plt.plot([1, 4, 6, 8], [3, 8, 3, 5])
plt.show()
```

```python
plt.plot(x, y)

plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law")

plt.xlim([20, 80])
plt.ylim([200, 800])

plt.show()
```

```python
x = np.linspace(0, 10, 20)
y = x * 2

plt.plot(x, y, "b+")

plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law")

plt.show()
```

```python
x = np.linspace(0, 5, 10)
y = np.linspace(3, 6, 10)

plt.plot(x, y, "r-", x, y**2, "b+", x, y**3, "g^")

plt.show()
```

```python
x = np.linspace(1, 10, 100)
y = np.log(x)

plt.figure(1)

plt.subplot(121)
plt.title("y = log(x)")
plt.plot(x, y)

plt.subplot(122)
plt.title("y = log(x)**2")
plt.plot(x, y**2)

plt.show()
```

```python
x = np.linspace(1, 10, 100)
y = x * 2

plt.figure(1)

plt.subplot(221)
plt.title("y = 2x")
plt.plot(x, y)

plt.subplot(222)
plt.title("y = log(2x)")
plt.plot(x, np.log(y))

plt.figure(2)

plt.subplot(221)
plt.title("y = (2x) ** 3")
plt.plot(x, y**3)

plt.subplot(222)
plt.title("y = e**(2x)")
plt.plot(x, np.e**y)

plt.show()
```
