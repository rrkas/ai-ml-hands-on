```python
btc = pd.read_csv("../../Data/crypto_data/bitcoin_price.csv")
ether = pd.read_csv("../../Data/crypto_data/ethereum_price.csv")
ltc = pd.read_csv("../../Data/crypto_data/litecoin_price.csv")
monero = pd.read_csv("../../Data/crypto_data/monero_price.csv")
neo = pd.read_csv("../../Data/crypto_data/neo_price.csv")
quantum = pd.read_csv("../../Data/crypto_data/qtum_price.csv")
ripple = pd.read_csv("../../Data/crypto_data/ripple_price.csv")
```

```python
def suffix(suff):
    return lambda x: str(x) + suff


btc.columns = btc.columns.map(suffix("_btc"))
ether.columns = ether.columns.map(suffix("_et"))
ltc.columns = ltc.columns.map(suffix("_ltc"))
monero.columns = monero.columns.map(suffix("_mon"))
neo.columns = neo.columns.map(suffix("_neo"))
quantum.columns = quantum.columns.map(suffix("_qt"))
ripple.columns = ripple.columns.map(suffix("_rip"))
```

```python
print(btc.shape)
btc.head()
```

```python
m1 = pd.merge(btc, ether, how="inner", left_on="Date_btc", right_on="Date_et")
m2 = pd.merge(m1, ltc, how="inner", left_on="Date_btc", right_on="Date_ltc")
m3 = pd.merge(m2, monero, how="inner", left_on="Date_btc", right_on="Date_mon")
m4 = pd.merge(m3, neo, how="inner", left_on="Date_btc", right_on="Date_neo")
m5 = pd.merge(m4, quantum, how="inner", left_on="Date_btc", right_on="Date_qt")
crypto = pd.merge(m5, ripple, how="inner", left_on="Date_btc", right_on="Date_rip")

print(crypto.shape)
crypto.head()
```

```python
close_cols = list(filter(lambda x: x.startswith("Close_"), crypto.columns))
curr = crypto[close_cols]
curr.head()
```

```python
sns.pairplot(curr)
plt.show()
```

```python
cor = curr.corr()
round(cor, 3)
```

```python
plt.figure(figsize=(10, 8))

sns.heatmap(cor, cmap="YlGnBu", annot=True)
plt.show()
```

# Case Study

```python
sns.set(style="white")
```

```python
market_df = pd.read_csv("../../Data/global_sales_data/market_fact.csv")
customer_df = pd.read_csv("../../Data/global_sales_data/cust_dimen.csv")
product_df = pd.read_csv("../../Data/global_sales_data/prod_dimen.csv")
shipping_df = pd.read_csv("../../Data/global_sales_data/shipping_dimen.csv")
orders_df = pd.read_csv("../../Data/global_sales_data/orders_dimen.csv")
```

```python
sns.boxplot(market_df["Sales"])
plt.yscale("log")
plt.show()
```

```python
df = pd.merge(market_df, product_df, how="inner", on="Prod_id")
print(df.shape)
df.head()
```

```python
sns.boxplot(x="Product_Category", y="Sales", data=df)
plt.yscale("log")
plt.show()
```

```python
sub_profit = df[(-1000 < df["Profit"]) & (df["Profit"] < 1000)]["Profit"]
sns.boxplot(x=df["Product_Category"], y=sub_profit)
plt.show()
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(121)
sns.boxplot(x=df["Product_Category"], y=df["Sales"])
plt.title("Sales")
plt.yscale("log")

plt.subplot(122)
sns.boxplot(x=df["Product_Category"], y=sub_profit)
plt.title("Profit")

plt.show()
```

```python
df = pd.merge(df, customer_df, how="inner", on="Cust_id")
print(df.shape)
df.head()
```

```python
sns.boxplot(x=df["Customer_Segment"], y=sub_profit)
plt.show()
```

```python
plt.figure(num=None, figsize=(12, 8), dpi=80, facecolor="w", edgecolor="k")
sns.boxplot(x=df["Customer_Segment"], y=sub_profit, hue=df["Product_Category"])
plt.show()
```

```python
sns.boxplot(x=df["Product_Category"], y=100 * df["Shipping_Cost"] / df["Sales"])
plt.ylabel("100 * (shipping cost / sales)")
plt.show()
```

```python
sns.barplot(x=df["Product_Category"], y=df["Sales"])
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(121)
sns.barplot(x=df["Product_Category"], y=df["Sales"])
plt.title("Average Sales")

plt.subplot(122)
sns.barplot(x=df["Product_Category"], y=df["Sales"], estimator=np.median)
plt.title("Median Sales")

plt.show()
```

# Time Series

```python
df = pd.merge(market_df, orders_df, how="inner", on="Ord_id")
print(df.shape)
df.head()
```

```python
df.info()
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"], dayfirst=True)
```

```python
df.info()
```

```python
time_df = df.groupby("Order_Date")["Sales"].sum()
time_df
```

```python
sns.lineplot(time_df)
plt.xticks(rotation=25)
plt.show()
```

```python
df["month"] = df["Order_Date"].dt.month
df["year"] = df["Order_Date"].dt.year

df.head()
```

```python
df_time = df.groupby(["year", "month"])["Sales"].mean()
df_time
```

```python
data = []
for k, v in df_time.items():
    data.append(
        {
            "year": k[0],
            "month": k[1],
            "mm-yyyy": f"{k[1]:02d}-{k[0]}",
            "value": v,
        }
    )
data = pd.DataFrame(data)
data
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x=data["mm-yyyy"], y=data["value"])
plt.xlabel("Time")
plt.xticks(rotation=90)
plt.ylabel("Sales")
plt.show()
```

```python
year_month = pd.pivot_table(df, values="Sales", index="year", columns="month")
year_month.head()
```

```python
plt.figure(figsize=(6, 6))

sns.heatmap(year_month, cmap="YlGnBu")
plt.show()
```
